# Optimization

```python
def fx(x):
    return 2*x -1

def gf(f):
    return (f - 3)**2

def compute_dg_dx(x):
    return 8*x - 16
```

```python
x = 6.0
lr = 0.1
print('g(f(x))=',gf(fx(x)))
```

```python
dgx = compute_dg_dx(x)
print(dgx)

x = x - lr*dgx
print(x)

print('g(f(x))=',gf(fx(x)))
```

```python
losses = []
for _ in range(20):
    dg_dx = compute_dg_dx(x)

    signal = gf(fx(x))
    losses.append(signal)

    x = x -lr*dg_dx
```

```python
# prompt: plot losses

import matplotlib.pyplot as plt

plt.plot(losses)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss over Iteration')
plt.show()
```
